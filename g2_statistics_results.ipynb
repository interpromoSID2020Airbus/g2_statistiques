### Study of scores per month

```python
# delete the years that do not contain a lot of information

mask = series_month['Date_Flown'] < '2015-06-01 00:00:00'
series_month = series_month.drop(series_month['Date_Flown'].loc[mask].index)
```

```python
# delete the years that do not contain a lot of information

mask = series_month['Date_Flown'] > '2019-09-01 00:00:00'
series_month = series_month.drop(series_month['Date_Flown'].loc[mask].index)
```

```python
# recovery of the average note over time for each type of note, 
# thanks to the function rate_per_date

overall = rate_per_date_all('Overall_Customer_Rating', series_month)
comfort = rate_per_date_all('Seat_Comfort', series_month)
food = rate_per_date_all('Food_And_Beverages', series_month)
money = rate_per_date_all('Value_For_Money', series_month)
service = rate_per_date_all('Cabin_Staff_Service', series_month)
```

```python
# Representation of notes over time, on the same chart

plt.figure(figsize=(20, 6), dpi=80)
p1 = plt.plot(overall['Date_Flown'], overall['mean'],
              label='Overall_Customer_Rating')
p2 = plt.plot(comfort['Date_Flown'], comfort['mean'],
              label='Seat_Comfort')
p3 = plt.plot(food['Date_Flown'], food['mean'],
              label='Food_And_Beverages')
p4 = plt.plot(money['Date_Flown'], money['mean'],
              label='Value_For_Money')
p5 = plt.plot(service['Date_Flown'], service['mean'],
              label='Cabin_Staff_Service')

plt.title("Evolution of grades per year")
plt.xticks(rotation='vertical')
plt.legend()
plt.show()
```

We observe that users are generally more severe over the years: all rates are falling.  
Moreover the overall rating is overall lower than the others.

#  Principal component analysies

The objective of this part is to perform PCAs (Principal Component Analysis) on the different variables with rates.
The goal is to cross these rates with an interesting variable to analyze.

Initially we used the variable *TYPE_OF_TRAVELLERS*, but once grouped, we had too few individuals.
Thus we chose to realize a PCA by grouping two variables: *TYPE_OF_TRAVELLERS* and *CABIN_CLASS* which allowed us to double the number of variables in the PCA.
We chose these variables because we found it interesting to know the rates that each group of travellers attributes.

Then we made a second PCA on the notes according to the "Airline_Name" variable of the *KAYAK_AIRLINE_RATE_GLOBAL* file.
We have selected only the companies with more than 1000 rates filled in for better results.
We chose this variable because we wanted to analyze the different airlines listed in this source and thus find out how they were distributed.

Finally, we performed a second PCA on the ratings in function of the airlines but this time on the ALL_DATA file but only with the AIRLINE QUALITY AIRLINE data source
We noted that the results were similar between these two PCAs.

We did other PCAs where we found results not relevant.

**Non relevant PCAs :**

Two PCAs with *TYPE_OF_TRAVELLERS* on full ALL_DATA and then only on the AIRLINE QUALITY AIRLINE data source.
We obtained results with only 4 individuals since we grouped the individuals by averaging the scores according to *TYPE_OF_TRAVELLERS* which is not much.

Two PCAs with *CABIN_CLASS* on the whole ALL_DATA source and then only on the AIRLINE QUALITY AIRLINE data source.
For the same reasons as the previous PCAs, we did not consider this PCA representative. 
We decided to keep only the first data source because we wanted to keep only the lines where the rates were given by the same people because all data sources
did not rate the same things. 

A PCA with *TYPE_OF_TRAVELLERS* on ALL_DATA_ANNOTATE.
We did not keep this PCA because it is only made on 6000 lines (airlines) which are then grouped in 4 as for the previous PCAs.

A PCA with *CONSTRUCTOR* on the ALL_DATA source where the *CONSTRUCTOR* column has been previously created. It informs the constructors of the aircraft in which the passengers travelled. 
To do this, we analyzed the names of the aircraft in the *AIRCRAFT_TYPE* column to deduce the constructor. 
After reflection, we realized that this last PCA is useless because only the rates concerning comfort are associated with the comfort. 
For example, *CABIN_STAFF_SERVICE* and *FOOD_AND_BEVERAGE* concern airlines and not constructors.

We then wanted to perform a PCA by grouping the variables into 4 groups according to airlines : Comfort, Food/Beverage, Staff, Entertainment/Equipment 
but once the groupings were performed we realized that the results were identical to the PCA on all scores. We just have more airlines.



## PCA on rates by companies with kayak rates

### Preprocessing

Selection of airline company with more than 1000 notes

```python
# index of these companies
indKayak = []
for i in range(kayak.shape[0]):
    if kayak['Number_Evaluation'].iloc[i] > 1000:
        indKayak.append(i)
```

Selection of notes columns

```python
kayakACP = kayak.iloc[indKayak]

# Airline name
label = kayakACP['Airline_Name']
kayakACP = kayakACP.iloc[:, 2:8]
```

Number of individuals

```python
n = kayakACP.shape[0]
n
```

Number of variables 

```python
p = kayakACP.shape[1]
p
```

Data normalisation

```python
sc = StandardScaler()
kayakCR = sc.fit_transform(kayakACP)
```

Instanciation 

```python
acpKayak = PCA(svd_solver='full')
```

Each individual will have the color corresponding to his mentions

```python
colors = kayak.iloc[indKayak]['Mention'].map(
    {"Bien": "red", "Correct": "yellow", 
     "Excellent": "blue", "M√©diocre": "green"}).values
```

### Calcul of coordinates

Coordinates

```python
coord = acpKayak.fit_transform(kayakCR)
```

Proportion explained variance

```python
acpKayak.explained_variance_ratio_
```

Eigen Values

```python
eigval = ((n-1)/n)*acpKayak.explained_variance_
eigval
```

```python
# Display of eigen values

plt.plot(np.arange(1, p+1), eigval)

plt.title("Display of eigen values")
plt.show()
```

The first principal component explains 95.03% of the total inertia.  
Moreover, there is a very significant break on the graph of the eigenvalues after the first axis.  
It is therefore the one on which we will base our analysis.  
So we decide to carry out our analyzes only with the first principal component.

### Display of Airline Companies in the first plan

```python
fig = plt.figure(figsize=(15, 15))

for i in range(n):
    x = coord[i, 0]
    y = coord[i, 1]
    plt.scatter(x, y, c=colors[i])
    plt.text(x, y, label.iloc[i])
    
plt.title("Display of Airline Companies in the first plan")
plt.show()
```

Quality of individuals corelations

```python
# Contribution of individuals to total inertia
di = np.sum(kayakCR**2, axis=1)
pd.DataFrame({'ID': kayakACP.index, 'd_i': di})

cos2 = coord**2
for j in range(p):
    cos2[:, j] = cos2[:, j]/di

pd.DataFrame(
    {'id': kayakACP.index, 'COS2_1': cos2[:, 0], 
     'COS2_2': cos2[:, 1]})
```

### Display of the correlation circle

Corelation of variables on the first plan

```python
sqrt_eigval = np.sqrt(eigval)

# Corelation of variables with axis
corvar = np.zeros((p, p))

for k in range(p):
    corvar[:, k] = acpKayak.components_[k, :] * sqrt_eigval[k]

# Coordinates of variables on the first plan
pd.DataFrame(
    {'id': kayakACP.columns, 'COR_1': corvar[:, 0], 
     'COR_2': corvar[:, 1]})
```

```python
fig, axes = plt.subplots(figsize=(10, 10))
axes.set_xlim(-1, 1)
axes.set_ylim(-1, 1)

for j in range(p):
    x = corvar[j, 0]
    y = corvar[j, 1]
    plt.scatter(x, y)
    plt.text(x, y, kayakACP.columns[j])

cercle = plt.Circle((0, 0), 1, color='blue', fill=False)
axes.add_artist(cercle)

plt.title("Display of the correlation circle")
plt.show()
```

### Interpretation

The first principal component is highly correlated with the 6 variables of our analysis.   
The graph of the Airline Companies shows an opposition between highly rated (left) and poorly rated companies (right).  

The variables are very little correlated to the second axis, we can still observe a slight opposition between the highly rated companies concerning on-board entertainment (up) and the highly rated companies concerning the cabin staff service and the boarding (down).   

In addition, the colours give us the mention that has been attributed to them: blue for the companies with the excellent mention, red for the companies with the good mention and yellow for those with the correct mention.
