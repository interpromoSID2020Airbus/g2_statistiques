# Shap introduction

In this notebook we are going to use several methods to explain the prediction models made, especially the shap method.

SHAP (SHapley Additive exPlanations) is a game theoretic approach to explain the output of any machine learning model.

The use of this method requires, for each model to be explained, a calculation of its shap values.
SHAP values represent a feature's responsibility for a change in the model output.  
We will use its values in order to be able to graphically explain the results found.   

In order not to weigh down the notebook, shap values are created by a function (see notebook functions) then stored in a csv file. Thus, we will start our analysis by importing these files directly.

## Interpretation of a shap summary plot

We will limit ourselves to the analysis of only 2 graphs:  
<p>
<li>a shap summary plot bar, which displays only the most explanatory features of the model.</li>
<li>a more complete shap summary plot to observe the effects of the explanatory features.</li>
</p>
The **shap.summary_plot** combines the importance of features with their effects. Each point in the summary plot represents an individual (a Shapley value for an individual). The position on the y-axis is determined by the variable to be explained and on the x-axis by the Shapley value. The color represents the value of the variable to be explained from bottom to top. The overlapping points are unstable in the direction of the y-axis, so we get an idea of the distribution of Shapley values by entity. Features are ranked according to their importance.

## Wifi

**Correlation between y_pred and features**

```python
df = wifi_Data["test_X_LR_wifi"].drop(
    wifi_Data["test_X_LR_wifi"].columns[0], axis="columns")
df["Y_pred"] = wifi_Model["LR_wifi"].predict(df)
ypredcorr = df.corr()[["Y_pred"]]
ypredcorr = ypredcorr.sort_values(by=['Y_pred'])
```

```python
plot_heatmap_2(ypredcorr.iloc[-20:-1], 15, 15)
plot_heatmap_2(ypredcorr.iloc[:20], 15, 15)
```

**Shap interpretation**

```python
test_X = wifi_Data["test_X_LR_wifi"].drop(
    wifi_Data["test_X_LR_wifi"].columns[0], axis="columns")
shap.summary_plot(shap_values_wifi, test_X, plot_type="bar")
```

```python
shap.summary_plot(shap_values_wifi.values, test_X)
```

**Interpretation**

The graphs allow us to know the words that had a significant impact on the predictions of the notes related to wifi.  
The graph shap allows us to refine this analysis and know how they have an impact.  

We notice that the less the word 'bad' is mentioned in the comments, the better the predictions of the scores related to wifi.  
On the other hand, the more the words 'clean', 'good', 'great', 'fast' and 'easy' are mentioned, the higher the predicted scores will be.  

As far as the word 'wifi' is concerned, the less it is mentioned the better the predictions will be, so users tend to talk about it when they are unhappy with it.

**Interpretation**

The prediction model of is a random forest model. These models allow us to achieve an explanation that is slightly different from the others. Here, the shap values are calculated for each of the predicted classes (-1, 0 and 1).  So we can analyze which features (word) explains the model for each class separately. We decided to analyze only the extreme classes (-1 and 1).  


For the predictions -1, i.e. users not satisfied with the waiting time :
- The presence of the words 'bad', 'hour', 'queue' and 'passenger' explain this prediction.
- The absence of the words 'terminal' and 'good' explain this prediction.

For prediction 1, it is to help users satisfied with the waiting time :
- The presence of the words 'good', 'clean' and 'friendly' explains this prediction.
- The absence of the words 'bad', 'passenger', 'hour' explain this prediction.
